```python
import os, sys
from KMeansGifer import KMeansGif
from matplotlib import pyplot as plt
import numpy as np
import fibo as f

image = plt.imread("kmeans.jpg")
k_offset = 4 # minimum k value never drop below 2
dims = image.shape
kmg = KMeansGif()
kmg.ydim, kmg.xdim, kmg.dpth = dims[0:3]
scaler = 1 #255.0
image = image / scaler
image_flat = image.reshape(dims[0] * dims[1], dims[2])
kari = [int(.01*f[0]*f[1]*f[2])%10+k_offset for f in image_flat]
kari = [v for i,v in enumerate(kari) if v != kari[i-1]]
```

```python
# Constant Delay w/wo Rotation
kmg.ResetImages()
kmg.imgIn = image_flat
kmg.doRotation = True
for i,k in enumerate(kari[:5]):
    kmg.n = k
    if True:    
        kmg.getImageKM()
    else:
        None#kmg.getImageDB() #not working-- what's the concept here-- think before executing
    kmg.AddImage()
kmg.SaveToDisk()
```

```python
# Single Image w/wo rotation
kmg.ResetImages()
kmg.imgIn = image_flat
kmg.doRotation = True
kmg.n = 12
kmg.getImageKM()
kmg.AddImage()
kmg.SaveToDisk()
```

```python
# With Fibo Delay No Rotation Option
kmg.ResetImages()
if True:
    procimg = np.log10(image_flat).astype('uint8')
else:
    procimg = image_flat
kmg.imgIn = procimg
kmg.doRotation = True
kd = kmg.getKsAndDelays(7,mink=2)
for kd_item in kd:
    kmg.n = kd_item[0]
    kmg.getImageKM()
    for d in range(kd_item[1]):
        kmg.AddImage()
kmg.images = kmg.images+kmg.images[-1::-1]
kmg.SaveToDisk()
    
```

```python
# With Breathe Delay No Rotation Option
kmg.ResetImages()
if True:
    procimg = np.log(image_flat).astype('uint8')
else:
    procimg = image_flat
kmg.imgIn = procimg
kmg.doRotation = False
kd = kmg.getKsAndBreathe(40,mink=2)
for kd_item in kd:
    kmg.n = kd_item[0]
    kmg.getImageKM()
    for d in range(kd_item[1]):
        kmg.AddImage()
kmg.images = kmg.images+kmg.images[-1::-1]
kmg.SaveToDisk()
```
